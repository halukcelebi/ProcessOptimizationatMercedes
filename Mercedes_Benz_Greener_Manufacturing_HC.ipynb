```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("train.csv")
x_test = pd.read_csv('test.csv')
```

```python
df
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop('ID', axis=1, inplace = True)
x_test.drop('ID', axis=1, inplace=True)
```

```python
df
```

```python
x_test
```

```python
y = df['y']
x = df.drop(columns=['y'])
x = pd.concat([x, x_test], ignore_index=True)
```

```python
x.shape, y.shape
```

```python
x = pd.get_dummies(x,drop_first=True)
```

```python
y_train=df['y']
```

```python
x_train = x.iloc[:4209, :]
x_test = x.iloc[4209:, :]
```

```python
x_train.shape, x_test.shape
```

```python
x_train
```

```python
x_test
```

```python

model=Sequential()
model.add(Dense(2048, activation="relu", input_dim=571))
model.add(Dense(1024, activation="relu"))
model.add(Dense(512, activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(1, activation="linear"))
```

```python
model.summary()
```

```python
 model.compile(loss='mean_absolute_error', # one may use 'mean_absolute_error' as alternative
                  optimizer='adam'
                                  )
```

```python
x_train = np.asarray(x_train).astype(np.float32)
y_train = np.asarray(y_train).astype(np.float32)
x_test = np.asarray(x_test).astype(np.float32)
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
x_test
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred1 = model.predict(np.asarray(x_train).astype(np.float32))
```

```python
# Calculate mean squared error
mse = mean_squared_error(y_train, y_pred1)

print(f"Mean Squared Error on Test Set: {mse}")
```
